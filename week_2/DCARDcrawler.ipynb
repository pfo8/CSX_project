<h3> Curl popular posts on https://www.dcard.tw/f/girl

```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
url = 'https://www.dcard.tw/_api/forums/girl/posts?popular=true'
res = requests.get(url)
data = json.loads(res.text)
```

```python
data = pd.DataFrame(data)
print(data)
```
